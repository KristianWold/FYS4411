# Analysis

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
localEnergies = np.loadtxt("energies.txt")
plt.plot(localEnergies)
plt.show()
```

```python
def smoothing(x, interval):
    x_smooth = np.zeros(x.shape)
    for i in range(len(x) - interval - 1):
        x_smooth[i] = np.mean(x[i:i+interval])
    return x_smooth

smooth = smoothing(localEnergies, 10000)
```

```python
plt.plot(smooth[:10000])
plt.show()
```

```python
def derivative(x):
    return x[1:] - x[:-1]
```

```python
plt.plot(derivative(smooth[:20000]))
plt.show()
```
